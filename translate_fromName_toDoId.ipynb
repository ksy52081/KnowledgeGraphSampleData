```python
# clustering 데이터 불러오기
clusteringData = pd.read_csv('ClusteringnAnalysis.csv', encoding='utf-8', sep=",")
clusteringData.head(10)
```

```python
# 코드 네임 데이터 불러오기
region_id_df = pd.read_csv('ID_DO_Mapping.csv', encoding='utf-8', sep=",")
region_id_df.head(5)
```

```python
import re
# 따옴표 제거하기
clusteringData['DoIds'] = clusteringData['DoIds'].apply(lambda x: re.findall(r"'(.*?)'", x))
clusteringData['ChildrenIds'] = clusteringData['ChildrenIds'].apply(lambda x: re.findall(r"'(.*?)'", x))

```

```python
clusteringData.head(5)
```

```python
for index, row in clusteringData.iterrows():
    values = row['DoIds']
    print(type(values))
```

```python
# # 빈 데이터프레임 생성
result_df = pd.DataFrame()


# 각 행의 Region 리스트를 반복하며 ID 값을 찾아서 새로운 데이터프레임에 추가
for index, row in clusteringData.iterrows():
    Ids = row['Id']
    ParentIds = row['ParentId']
    ChildrenIdsList = row['ChildrenIds']
    # DoIdList = ast.literal_eval(row['DoIds'].replace("'", ''))  # 따옴표를 제거하고 문자열을 리스트로 파싱
    
    # DoIdList = ast.literal_eval(row['DoIds'])  # 문자열을 리스트로 파싱
    DoIdList = row['DoIds']
    # if len(DoIdList) == 0:
        # continue  # regions가 비어 있는 경우에 대한 처리

    levels = row['level']
    
    # 각 Region을 ID로 변환
    numericids = [region_id_df.loc[region_id_df['Region'] == DoIds, 'ID'].values[0] for DoIds in DoIdList]
    
    # 새로운 데이터프레임에 ID와 Value 추가
    row_data = {'Id': Ids, 'ParentId': ParentIds, 'ChildrenIds':ChildrenIdsList,'DoIds':numericids,'level':levels}
    # row_data = {'Id': Ids, 'ParentId': str(ParentIds), 'ChildrenIds':str(ChildrenIdsList),'DoIds':str(numericids),'level':levels}
    print(row_data)
    temp_df = pd.DataFrame(row_data)
    print(temp_df)
    result_df = pd.concat([result_df, temp_df])
    print(result_df)
    

# 결과를 출력
print(result_df)
```

```python
# 'region_id_df'에서 Region과 ID 값만 선택
region_id_mapping = region_id_df[['ID', 'Region']]

# Region 값을 리스트로 변환하고 빈 리스트는 NaN으로 처리
# clusteringData['DoIds'] = clusteringData['DoIds'].apply(lambda x: x.strip('[]').split(', ') if x != '[]' else pd.NA)

# Region 값이 리스트인 행을 분리하여 새로운 데이터프레임 생성
clusteringData = clusteringData.explode('DoIds')

# 'df'와 'region_id_mapping'을 병합하여 Region 값을 ID로 변환
# result_df = clusteringData.merge(region_id_mapping, left_on='Region', right_on='Region', how='left')
result_df = pd.merge(clusteringData, region_id_mapping, left_on = 'DoIds', right_on = 'Region', how='left')
```

```python
#  EXAMPLE

import pandas as pd

# 예시 데이터프레임 생성
data = {
    'Region': ['[지역1, 지역3]', '[지역2]', '[]', '[지역1, 지역2]', '[지역3]', '[]']
}

df = pd.DataFrame(data)

# 'region_id_df'에서 Region과 ID 값만 선택
region_id_mapping = region_id_df[['ID', 'Region']]

# Region 값을 리스트로 변환하고 빈 리스트는 NaN으로 처리
df['Region'] = df['Region'].apply(lambda x: x.strip('[]').split(', ') if x != '[]' else pd.NA)

# Region 값이 리스트인 행을 분리하여 새로운 데이터프레임 생성
df = df.explode('Region')

# 'df'와 'region_id_mapping'을 병합하여 Region 값을 ID로 변환
result_df = df.merge(region_id_mapping, left_on='Region', right_on='Region', how='left')

# 결과를 출력
print(result_df)
```
